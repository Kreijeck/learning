```python
%matplotlib inline

import cv2
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread("bild.jpg")
```

```python
# RGB ist vertauscht, da openCV BGR verwendet
plt.imshow(img)
plt.show()
```

```python
i = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()
```

# Von Matplotlib wird RGB verwendet, OPENCV verwendet BGR

```python
print(img[0][0])
print(i[0][0])
```

```python
g = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(g, "gray")
plt.show()
```

```python
# typ der zahlen eines Arrays kann mit dtype definiert werden
r = np.array([1, 2, 3, 4], dtype="uint8")
# funktioniert noch
r = r + 250
print(r)
# sprengt wertebereich, damit fängt man wieder von 0 an zu zählen
r = r + 100
print(r)
```

```python
i = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()
```

```python
# überall wo die grenze überschritten wird gibt es ganz komische Farben
increased = img + 50
i = cv2.cvtColor(increased, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()
```

```python
z = np.zeros((4608, 3456, 3), dtype="uint8") + 50

# wenn bei Addition größer als 255, bleibt es auf 255!
increased = cv2.add(img, z)
print("mit überlauf")
print(img[1000][0])
print(z[1000][0])
print(increased[1000][0])
print("ohne überlauf")
print(img[0][0])
print(z[0][0])
print(increased[0][0])
```

```python
# Original
print("Original")
i = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()

#heller
print("Heller")
i = cv2.cvtColor(increased, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()

```

# Rechteck einzeichnen

```python
img = cv2.imread("bild.jpg")

# Parameter: [image-cv2], [POS1], [POS2], [BGR], [Dicke]
# POS1 und POS2 müssen 2 diagonale Punkte in einem Rechteck sein
cv2.rectangle(img, (200, 200), (1500, 1500), (0,0,255), 50)

i = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()

```
